```python
df6=pd.read_csv('/content/data_processors.csv')
```

```python
df6.rename(columns={' Processor                  ':'processor_name',' Rating ':'rating'},inplace=True)
```

```python
df6['processor_name']=df6['processor_name'].str.strip()
```

```python
df
```

```python
df=df.merge(df6,how='left',on='processor_name')
```

```python
df=df.rename(columns={'rating_y':'processor_score'})
```

```python
df['processor_score'] = df.groupby('pricewise_cat')['processor_score'].transform(
    lambda x: x.fillna(x.mean())
)
```

```python
df['processor_score']=np.round(df['processor_score'],1)
```

```python
df['ram_capacity(GB)'].value_counts()
```

```python
df['ram_capacity(GB)']=df['ram_capacity(GB)'].astype('string')
```

```python
ram_scores={'8':10 ,
'16.0':9,
'12.0':8.5,
'8.0':7.5,
'6.0':6.5,
'4.0':5.5,
'3.0':4.5,
'2.0':3.5,
'1.0':2}
df['ram_scores']=df['ram_capacity(GB)'].map(ram_scores)
```

```python
df['ram_capacity(GB)']=df['ram_capacity(GB)'].astype('float')
```

```python
df[df['rom_capacity(GB)']==1]
```

```python
df['rom_capacity(GB)']=df['rom_capacity(GB)'].astype('string')
```

```python
df['rom_capacity(GB)']=df['rom_capacity(GB)'].astype('string')
```

```python
rom_score={
    '1':10,
    '512':9,
    '256':8,
    '128':7,
    '64':5.5,
    '32':4,
    '16':2.5
}
df['rom_score']=df['rom_capacity(GB)'].map(rom_score)
```

```python
pd.set_option('display.max_columns', 1000);
```

```python
df['rom_capacity(GB)']=df['rom_capacity(GB)'].astype('int')
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.reset_index(inplace=True)
```

```python
df.drop(columns=['index'],inplace=True)
```

```python
df.info()
```

```python
df['normalized_refresh_rate'] = df.groupby('pricewise_cat')['normalized_refresh_rate'].transform(
    lambda x: x.fillna(x.mean())
)
```

```python
df['ram_scores'] = df.groupby('pricewise_cat')['ram_scores'].transform(
    lambda x: x.fillna(x.mean())
)
```

```python
df['has_FC'].fillna(0,inplace=True)
```

```python
df['performance_index'] = (
    0.35 * df['processor_score'] +
    0.25 * df['ram_scores'] +
    0.10 * df['rom_score'] +
    0.15 * df['normalized_refresh_rate'] +  # Already normalized
    0.05 * df['has_FC'] +
    0.10 * (df['has_5g'] + df['has_nfc']) / 2  # Average of 5G and NFC
)
```

```python
df['normalized_refresh_rate']=np.round(df['normalized_refresh_rate'],1)
```

```python
df['performance_index']=np.round(df['performance_index'],2)
```

```python
df.to_csv('good_smartphones.csv')
```
