```python
import numpy as  np
import pandas as pd
```

```python
df=pd.read_csv('/content/cleaned_smartphone_dataset.csv')
```

```python
df.head()
```

```python
df.drop(columns='Unnamed: 0',inplace=True)
```

```python
df
```

```python
df['processor_cores'].value_counts()
```

```python
df['processor_cores']=df['processor_cores'].str.replace('Octa','8')
df['processor_cores']=df['processor_cores'].str.replace('Hexa','6')
df['processor_cores']=df['processor_cores'].str.replace('Quad','4')
df['processor_cores']=df['processor_cores'].str.replace('Deca','10')
df['processor_cores']=df['processor_cores'].str.replace('Nine','9')
df['processor_cores']=df['processor_cores'].str.replace('Dual','2')
```

```python
df['processor_cores'].value_count()
```

```python
df['processor_cores']=df['processor_cores'].astype('int')
```

```python
df
```

```python
pd.set_option('display.max_columns', 1000);
```

```python
df
```

```python
df['processor_speed'].value_counts()
```

```python
df['processor_speed']=df['processor_speed'].str.replace('2MHz','0.2GHz')
```

```python
df[df['processor_speed']=='0.2GHz']
```

```python
df['processor_speed']=df['processor_speed'].str.strip().str.split("G").str.get(0)
```

```python
df.rename(columns={'processor_speed':'processor_speed(GHz)'},inplace=True)
```

```python
df['processor_speed(GHz)']=df['processor_speed(GHz)'].astype('float')
```

```python
df
```

```python
df['number_rear_camera'].value_counts()
```

```python
df['number_rear_camera']=df['number_rear_camera'].str.replace('Triple','3')
df['number_rear_camera']=df['number_rear_camera'].str.replace('Dual','2')
df['number_rear_camera']=df['number_rear_camera'].str.replace('Quad','4')
```

```python
df['number_front_cameras'].value_counts()
```

```python
df['number_front_cameras']=df['number_front_cameras'].str.replace('Single','1')
df['number_front_cameras']=df['number_front_cameras'].str.replace('Dual','2')
```

```python
df['front_camera_resol']=df['front_camera_resol'].str.strip().str.split('M').str.get(0)
```

```python
df['front_camera_resol']=df['front_camera_resol'].astype('float')
```

```python
rear_primary_resol=df['rear_camera_resol'].str.strip().str.split('M').str.get(0)
```

```python
df.insert(21,'rear_primary_resol(MP)',rear_primary_resol)
```

```python
df['rear_primary_resol(MP)']=df['rear_primary_resol(MP)'].astype('float')
```

```python
k=df['battery_capacity(mAh)'].mean()
k=np.round_(k,decimals=2)
```

```python
df['battery_capacity(mAh)'].fillna(k,inplace=True)
```

```python
p=np.round(df['rating_FC(W)'].mean(),decimals=0)
p
```

```python
df['rating_FC(W)']=df['rating_FC(W)'].fillna(p)
```

```python
df
```

```python
df.to_csv('cleaned(1)_smartphone_dataset.csv')
```
